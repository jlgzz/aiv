```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pic= cv2.imread("flor.jpg")
pic= cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
```

```python
scale=0.1
new_width = int(pic.shape[1]*scale)
new_heigh= int(pic.shape[0]*scale)
pic_resized=cv2.resize(pic,(new_heigh,new_width))
plt.imshow(pic_resized)
```

```python
gray_pic= cv2.cvtColor(pic_resized,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_pic, cmap='gray')

```
